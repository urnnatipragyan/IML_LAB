```python
import numpy as np

class LogisticRegression:
    def __init__(self, learning_rate=0.01, n_iters=1000):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_iters):
            linear_model = np.dot(X, self.weights) + self.bias
            y_predicted = self.sigmoid(linear_model)

            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y))
            db = (1 / n_samples) * np.sum(y_predicted - y)

            self.weights -= self.lr * dw
            self.bias -= self.lr * db

    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        y_predicted = self.sigmoid(linear_model)
        return [1 if i > 0.5 else 0 for i in y_predicted]



n_samples = int(input("Enter number of samples: "))
n_features = int(input("Enter number of features: "))

X = []
y = []

print("\nEnter feature values row by row:")

for i in range(n_samples):
    row = list(map(float, input(f"Sample {i+1} (space-separated {n_features} values): ").split()))

    if len(row) != n_features:
        raise ValueError("Incorrect number of features entered.")

    X.append(row)

print("\nEnter labels (0 or 1):")
for i in range(n_samples):
    label = int(input(f"Label for sample {i+1}: "))

    if label not in [0, 1]:
        raise ValueError("Labels must be 0 or 1.")

    y.append(label)

X = np.array(X)
y = np.array(y)

model = LogisticRegression(learning_rate=0.1, n_iters=1000)
model.fit(X, y)

print("\nEnter a new sample to predict:")
test_sample = list(map(float, input(f"Enter {n_features} feature values: ").split()))

if len(test_sample) != n_features:
    raise ValueError("Incorrect number of features.")

test_sample = np.array([test_sample])

prediction = model.predict(test_sample)

print("\nPrediction:", prediction[0])
```
